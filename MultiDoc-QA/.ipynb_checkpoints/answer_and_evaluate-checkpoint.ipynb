```python
import json
import argparse
import collections
import numpy as np
import os
import re
import string
import sys
import unicodedata
```

```python
import openai

from transformers import GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
```

```python
def generate(prompt,max_tokens=256, temperature=0):
    response = openai.Completion.create(
        engine="text-davinci-002",
        prompt=prompt,
        temperature=temperature,
        max_tokens=max_tokens,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0
    )

    return response["choices"][0]['text']

def normalize_answer(s):
  """Lower text and remove punctuation, articles and extra whitespace."""
  def remove_articles(text):
    regex = re.compile(r'\b(a|an|the)\b', re.UNICODE)
    return re.sub(regex, ' ', text)
  def white_space_fix(text):
    return ' '.join(text.split())
  def remove_punc(text):
    exclude = set(string.punctuation)
    return ''.join(ch for ch in text if ch not in exclude)
  def lower(text):
    return text.lower()
  def remove_accents(input_str):
      nfkd_form = unicodedata.normalize('NFKD', input_str)
      only_ascii = nfkd_form.encode('ASCII', 'ignore')
      return only_ascii.decode("utf-8")

  return white_space_fix(remove_articles(remove_punc(lower(remove_accents(s)))))

def compute_f1(a_gold, a_pred):
  gold_toks = get_tokens(a_gold)
  pred_toks = get_tokens(a_pred)
  common = collections.Counter(gold_toks) & collections.Counter(pred_toks)
  dif_words = list(set(gold_toks) - set(pred_toks))
  if dif_words == ['years']:
      return 1
  num_same = sum(common.values())
  if len(gold_toks) == 0 or len(pred_toks) == 0:
    # If either is no-answer, then F1 is 1 if they agree, 0 otherwise
    return int(gold_toks == pred_toks)
  if num_same == 0:
    return 0
  precision = 1.0 * num_same / len(pred_toks)
  recall = 1.0 * num_same / len(gold_toks)
  f1 = (2 * precision * recall) / (precision + recall)
  return f1


  return white_space_fix(remove_articles(remove_punc(lower(remove_accents(s)))))

def get_tokens(s):
  if not s: return []
  return normalize_answer(s).split()

def compute_exact(a_gold, a_pred):
  gold_toks = get_tokens(a_gold)
  pred_toks = get_tokens(a_pred)
  dif_words = list(set(gold_toks) - set(pred_toks))
  if dif_words == ['years']:
      return 1
  else:
      return int(normalize_answer(a_gold) == normalize_answer(a_pred))
```

```python
list_prompt_answer = pd.read_csv('./data/list_prompt_answer_ground_truth.csv')
list_prompt_answer = list_prompt_answer.drop(columns=['Unnamed: 0'])
list_prompt_answer.iloc[:1]
```

```python
STATIC_PROMPT = """Follow examples bellow and generate value for Evidence and value for Answer. For each example, use the documents to provide an answer to the question and cite evidence from the documents to support the answer. All questions have an answer, and just one answer. Please complete with valid format. It must have 2 fields Evidence and Answer. If the answer is a value, complete with just one word for the unit used. Example: prefer "5 years" than just 5 or "5 years old". """
```

```python
%%time
temperature = 0
attempts = 1
pattern = "(?<=Answer:)(.*)$"
new_list_prompt_answer = pd.DataFrame({})
for _, item in tqdm(list_prompt_answer.iterrows()):
    item['responses'] = []
    item['results'] = ''
    item['completions'] = []
    item["asked_twice"] = None
    for i in range(attempts):
        prompt = STATIC_PROMPT
        text = f"\nDocument {1}: Title: {item['passage']}. Content: {item['text']}\n"
        prompt+= text
        prompt += f"\nQuestion: Based on the above documents, {item['question']}\n\nEvidence:"

        res = generate(prompt,temperature=temperature)
        if "Answer" not in res:
            item["new_prompt"] = "{0}{1}\n\nAnswer:".format(prompt, res)
            res2= generate(item["new_prompt"])
            item['results'] = "{0}\n\nAnswer: {1}".format(res, res2)
            item['responses'].append(res2)
            item['completions'].append("{0}\n\nAnswer: {1}".format(res, res2))
            item["asked_twice"] = True
        else:
            matches = re.findall(pattern, res)
            if len(matches) > 0: 
                response = matches[0]
                item['responses'].append(response)
            item['results'] = res 
            item['completions'].append(res)
            item["asked_twice"] = False
        new_list_prompt_answer = pd.concat((new_list_prompt_answer,item.to_frame().T))
new_list_prompt_answer.to_csv("./data/list_prompt_answer_ground_truth_with_resp.csv")
```

```python
new_list_prompt_answer['answer'] = new_list_prompt_answer['answer_value']
```

```python
f1s = []
ems = []
lista_resposta_llm = []
lista_type_answer = []

for _, item in tqdm(new_list_prompt_answer.iterrows()):
    response =  normalize_answer(item['results'].split("Answer:")[1].strip().replace('\n',''))

    f1 = compute_f1(item['answer'], response)
    f1s.append(f1)
    ems.append(compute_exact(item['answer'], response))
    if item['answer'] in ['yes', 'no']:
       type_answer = 'binary'
    else:
       type_answer = 'value'
    lista_type_answer.append(type_answer)    
    lista_resposta_llm.append(response)

print("\nF1:",np.mean(f1s))
print("EM:",np.mean(ems))
```
